Now with drink ID's from above query, we can look up each drink, get the ingredients in it, and append the ingredient list to the drink dictionary above.

```python
lookupEndpoint = "lookup.php?i="

def drinkByID(id):
    data = requests.get(baseUrl + lookupEndpoint + id)
    return json.loads(data.text)
```

```python
def getIngreds(drinkId):
    d = drinkByID(drinkId)
    ingreds = []
    for i in range(1,16):  # API has 16 fields for ingredients of each drink, most of them empty/None
        s = "strIngredient" + str(i)
        if not d['drinks'][0][s]:
            break
        ingreds.append(d['drinks'][0][s])
    return ingreds
```

```python
for d in vodkaDrinks['drinks']:
    dID = d['idDrink']
    d['ingreds'] = getIngreds(dID)
```
